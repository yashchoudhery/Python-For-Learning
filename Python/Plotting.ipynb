```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3]
y=[4,5,6]

plt.plot(x,y)
plt.show()   #for showing or printing the graph
```

```python
plt.plot(x,y,marker="o",linestyle=" ")
plt.show() 
```

```python
plt.plot(x,y,marker="<",linestyle='-.',color='blue')  #marker=o,<,>,^,1 etc 
plt.show() 
```

```python
plt.plot(x,y,marker="<",linestyle='-.',color='blue',ms=30,mfc='green',mec='orange',linewidth=3)
plt.show()
```

```python
x=np.arange(5,25)
x
```

```python
plt.scatter(x,x**2,color='r',label='x^2')
plt.scatter(x,x**3,color='b',label='x^3')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
plt.scatter(x,x**2,color='r',label='x^2',marker='^')
plt.scatter(x,x**3,color='g',label='x^3',marker='o')
plt.scatter(x,x**4,color='g',label='x^4',marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot')
plt.legend()
plt.grid()
plt.show()
```
